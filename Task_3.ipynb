## Importing necessities

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing the data from file 
code = pd.read_csv('SampleSuperstore.csv')
code
```

```python
#displaying the heading data
code.head(5)
```

```python
#displaying the ending data
code.tail(5)
```

## Analyzing  and observing the Data 


```python
#analyzing the code mathematically
code.describe()
```

```python
#observing the data info 
code.info()
```

## Removing dupilcates rows from the data 

```python
#checking for duplicate rows
print("Number of duplicate rows =",code.duplicated().sum())
```

```python
#removal of duplicate rows 
code.drop_duplicates(inplace=True)
code
```

## plotting the data 

```python
#displaying the data through histogram plots 
code.hist(bins = 50 ,figsize=(15,10))
plt.show
```

```python
# comparing categories with sub-categories
plt.figure(figsize = (10, 5))
plt.bar('Sub-Category', 'Category', data=code, color='black')
plt.title('Category vs Sub-Category', size=25)
plt.xticks(rotation=70 ,size=15)
plt.yticks(size=15)
plt.show()
```

## Counting total number of states 

```python
# counting the total number of states
print("Total states are ",code['State'].nunique())
print(code['State'].value_counts())
```

## Profits in States 

```python
# sorting the proifts in states in decreasing order
states_profit = code.groupby(by="State").sum()['Profit'].sort_values(ascending=False).reset_index()
```

```python
#plotting the data of top 10 states
plt.figure(figsize=(10,5))
plt.title('States with maximum profit',size=25)
sns.barplot(x='State',y='Profit',data=states_profit.head(15))
plt.xticks(rotation=90,size=15)
plt.yticks(size=15)
plt.xlabel('States',size=20)
plt.ylabel('Profits',size=20)
plt.show()
```

```python
#plotting data of states with least profit 
plt.figure(figsize = (10,5))
plt.title('States with Minimum Profits', size=25)
sns.barplot(x='State', y='Profit', data=states_profit.tail(15)) 
plt.xticks(rotation=60, size=15)
plt.yticks(size=15)
plt.xlabel('States', size=20)
plt.ylabel('Profits', size=20)
plt.show()
```

## Profits in Cities 

```python
#sorting profit of cities in decreasing order 
cities_profit = code.groupby(by="City").sum()['Profit'].sort_values(ascending=False).reset_index()
```

```python
#plotting data of 10 cities with highest data 
plt.figure(figsize = (10,5))
plt.title('Cities with maximum profit',size=25)
sns.barplot(x='City',y='Profit',data=cities_profit.head(15))
plt.xticks(rotation=60,size=15)
plt.yticks(size=15)
plt.xlabel('Cities',size=15)
plt.ylabel('Profits',size=15)
plt.show()
```

```python
#plotting the data of bottom 10 cities with least profits
plt.figure(figsize = (10,5))
plt.title('Cities with Minimum Profits', size=25)
sns.barplot(x='City', y='Profit', data=cities_profit.tail(15)) 
plt.xticks(rotation=60, size=15)
plt.yticks(size=15)
plt.xlabel('City', size=20)
plt.ylabel('Profits', size=20)
plt.show()
```

## Category Plots 

```python
sns.countplot(x='Category', data=code)
plt.title('Category counts', size=20);
```

### It seems technology has least supply while office supplies have the most 

```python
#profits in term of Categories
category_profit = code.groupby('Category').sum()['Profit'].sort_values(ascending = False)

plt.figure(figsize=(6,6))
plt.title('Profits in terms of Categories', size=20)
category_profit.plot(kind='bar',color='red')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()

```

### And the most profitable category is technology

```python
#plot sales vs profit 
sales = code.groupby('Category').sum()['Sales'].nlargest(n=10)
#top 10 states with highest sales
profits = code.groupby('Category').sum()['Profit'].nlargest(n=10)
#top 10 states with highest profits
plt.style.use('seaborn')
sales.plot(kind='bar',fontsize=10,color='black')
profits.plot(kind='bar',fontsize=10,color='m')
plt.xlabel('Categories',fontsize=10)
plt.ylabel("Total states and Profits")
plt.title("Sales vs Profit in terms of Categories",size = 10)
plt.show()
```

## Sub-category Plots 

```python
plt.figure(figsize = (17,8))
sns.countplot(x='Sub-Category',data=code)
plt.title('Sub-Category counts',size=20)
```

```python
#profits of sub-categories 
sub_profit = code.groupby('Sub-Category').sum()['Profit'].sort_values(ascending = False)
plt.figure(figsize=(16,6))
plt.title('Profits in terms of Sub-Categories',size=15)
sub_profit.plot(kind='bar',color='black')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
#plotting for losses of sub-category
min_profit = code.groupby('Sub-Category').min()["Profit"].sort_values()
plt.figure(figsize = (8,6))
min_profit.plot()
plt.title('Sub-Categories with Highest Losses',size=20)
plt.xticks(rotation=50,fontsize=15,color='red')
plt.yticks(fontsize=15)
plt.show()
```

## Shipping mode plottings

```python
sns.countplot(x='Ship Mode',data=code)
plt.title("Shipping Mode counts",size=20);
```

### It is seen that shipping mode is very profitable compared to other modes .

```python
#plotting profits in shipping mode
ship_profit = code.groupby('Ship Mode').sum()['Profit'].sort_values(ascending = False)
plt.figure(figsize=(16,6))
plt.title('Profits in terms of Shipping Modes',size=20)
ship_profit.plot(kind='bar',color='red')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
ship_profit_avg = code.groupby('Ship Mode').mean()['Profit'].sort_values(ascending = False)

plt.figure(figsize=(16,6))
plt.title('Average Profits in terms of Shipping Modes', size=20)
ship_profit_avg.plot(kind='bar',color='green')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
#discounts in shipping mode
ship_discount = code.groupby('Ship Mode').sum()['Discount'].sort_values(ascending = False).reset_index()

plt.figure(figsize=(10,6))
plt.title('Discounts in terms of Shipping Modes', size=20)
sns.barplot(x='Ship Mode', y='Discount', data=ship_discount)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

## Segment Plots

```python
sns.countplot(x="Segment",data=code)
plt.title('Segment counts',size=20);
```

### Consumer orders are the most while office orders are the least.

```python
segment_profit = code.groupby('Segment').mean()['Profit'].sort_values(ascending = False).reset_index()

plt.figure(figsize=(8,5))
plt.title('Profits according to Segments', size=20)
sns.barplot(x='Segment', y='Profit', data=segment_profit)
plt.xticks(rotation=60 , fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
sales = code.groupby('Segment').mean()['Sales'].sort_values(ascending = False)
profits = code.groupby('Segment').mean()['Profit'].sort_values(ascending = False)

plt.style.use('seaborn')
sales.plot(kind='bar', fontsize=15, color='black')
profits.plot(kind='bar', fontsize=15, color='m')
plt.xlabel('Segments',fontsize=15)
plt.ylabel('Total Sales and Profits', fontsize=15)
plt.title("Sales vs Profits in terms of Segments", size=20)
plt.show()
```

## Region Plots 

```python
sns.countplot(x='Region', data=code)
plt.title('Region counts', size=20);
```

```python
#profits in region
regions_profit = code.groupby('Region').sum()['Profit'].sort_values(ascending = False).reset_index()

plt.figure(figsize=(10,5))
plt.title('Profits according to Regions', size=20)
sns.barplot(x='Region', y='Profit', data=regions_profit)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
#plot of sales vs profts in terms of region
sales = code.groupby('Region').sum()['Sales'].sort_values(ascending = False)
profits = code.groupby('Region').sum()['Profit'].sort_values(ascending = False)

plt.style.use('seaborn')
sales.plot(kind='bar', fontsize=15, color='black')
profits.plot(kind='bar', fontsize=15, color='m')
plt.xlabel('Regions', fontsize=15)
plt.ylabel('Total Sales and Profits', fontsize=15)
plt.title("Sales vs Profits in terms of Regions", size=15)
plt.show()
```

### Its clear seen that western zones and eastern zones are the most dominant in term of sales as well as profits. 

## Sales,Profits and Discounts 

```python
#profits vs discount
plt.figure(figsize=(10,7))
sns.lineplot(x='Discount',y='Profit',data=code,label='Discount',color='black')
plt.title("Profit vs Discount",size=15)
plt.legend()
plt.show()
```

## The plot shows direct proportional relation between profits and discount.

## Profits and Sales have a moderately strong correlation

```python
#profits for sales vs profits in top 10 states

#top 10 states wit highest sales 
sales = code.groupby('State').sum()['Sales'].nlargest(n=15)
#top 10 states with highest profits
profits = code.groupby('State').sum()['Profit'].nlargest(n=15)

sales.plot(kind='bar',fontsize=10,color='black')
profits.plot(kind='bar',fontsize=10,color='m')
plt.xlabel('States',fontsize=15)
plt.ylabel('Total sales and profits',fontsize=15)
plt.title('Sales vs Profit in States',fontsize=15)
plt.show()
```

```python
#plot for sales vs profits with discount
plt.style.use('seaborn')
code.plot(kind="scatter", x="Sales", y="Profit", c="Discount", s=20, fontsize=16, colormap="Spectral")
plt.ylabel('Total Profits', fontsize=15)
plt.title("Sales vs Profits on Discounts", fontsize=20)
plt.show()
```
